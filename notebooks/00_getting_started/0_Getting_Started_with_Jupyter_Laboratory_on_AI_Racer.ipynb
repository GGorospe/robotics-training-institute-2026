## Reading the Execution Counter
## Look at the brackets to the left of the code cell you just ran, they should now show a number, like `[1]`.
## That number tells you **the order in which cells have been run** — not the order they appear in the notebook, but the actual order *you* ran them in.

### A few things worth knowing:
- `[ ]` (empty brackets) means a cell has never been run.
- `[*]` means the cell is **currently running** — wait for it to finish before running another cell that depends on it.
- `[7]` means this was the 7th cell you ran, in whatever order you ran things.

## Running Cells Out of Order (and Running Them Twice)
## This is one of the most important things to understand about Jupyter notebooks: **the notebook doesn't care what order the cells appear on the page — it only remembers what order you actually ran them in.**
## This means it's possible to run cells out of order, or run the same cell more than once, and end up with a confusing result if you're not paying attention.

## Run the two cells below **in order**, then run the second one a few more times by itself and watch what happens.

````python
# Run this cell once to start the counter at 0
counter = 0
print(counter)
````

````python
# Now run THIS cell by itself, several times in a row.
# Notice the number keeps going up, even though the code looks the same every time!
counter = counter + 1
print(counter)
````

## Reading Errors Without Panicking
## Everyone runs into Python errors — it happens to experienced programmers every day, not just beginners. The important skill isn't avoiding errors, it's knowing how to read one.

## Let's trigger one on purpose. Run the cell below.

````python
# This cell has a typo on purpose, it will produce an error when you run it.
print(counnter)
````

## What Just Happened?
## That red block of text is called a **traceback**. It looks intimidating, but it's actually just Python trying to help you.

### The strategy: **read the last line first.**
- The last line tells you the actual error type and message — in this case, something like `NameError: name 'counnter' is not defined`.
- That's Python telling us it has no idea what `counnter` is, because we misspelled `counter`.
- The lines above the last one show *where* in the code the error happened — useful once you're comfortable, but the last line is where to start.
